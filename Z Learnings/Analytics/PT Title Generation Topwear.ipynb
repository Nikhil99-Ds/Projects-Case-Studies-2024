````
few_short = '''
Example 1
Specifications:
{'Color': 'Black',
'Pattern': 'Dots',
'Length': 'Ankle Length',
'Occasion': 'Formal',
'Gender': 'Women',
'Fit': 'Slim Fit',
'Product Type': 'Treggings'}
Title:
Women's Black Dotted Ankle Length Slim Fit Formal Treggings

Example 2:
Specifications:
{'Color': 'Navy',
'Pattern': 'Solid',
'Sleeves': 'Sleeveless',
'Neck' : 'Round Neck',
'Occasion': 'Formal',
'Gender': 'Women',
'Fit': 'Slim Fit',
'Product Type': 'Dress'}
Title: 
Navy Solid Sleeveless Round Neck Women's Formal Slim Fit Dress

Example 3:
Specifications:
{'Color': 'White',
'Pattern': 'Print',
'Length': 'Calf-Length',
'Sleeves': 'Sleeveless',
'Neck' : 'Round Neck',
'Occasion': 'Formal',
'Gender': 'Women',
'Fit': 'Slim Fit',
'Product Type': 'Kurta Set'}
Title: 
Formal White Print Round Neck Sleeveless Women Calf-Length Slim Fit Kurta Set
'''
````

````
for iteration in range(100):
      features = P_json[iteration]
      del P_json[iteration]['ProductID']
      del P_json[iteration]['Brand']
      prompt = f'''
                Craft a compelling product description of a garment up to 20 words using the provided specifications in a line.
                Use only usefull words from the Specifications.
                Only inspire from the example.
                Specifications:
                ```{features}```
                Example:
                ```{few_short}```
                '''

    # OpenAI Chat API request
      response = openai.ChatCompletion.create(
                engine="abfrlgpt",
                temperature = 0,
                messages=[
                    # Creative content writer/creative copy writer for a premium fasion retail company.
                    {"role": "system", "content": "Assistant for a premium fashion retail company, tasked with creating website product title."},
                    {"role": "user", "content": prompt}
                ]
            )
      
      output = response['choices'][0]['message']['content']
      if output:
          generated_content = output
          Product_json[iteration]['Title'] = output
````
